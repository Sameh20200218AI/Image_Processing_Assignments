```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = '/content/Al-Hilal-Neymar-Photo-Wallpaper.jpg'  
image = cv2.imread(image_path)

# Resize the image
target_height, target_width = 1000, 1000  # Specify the target height and width for resizing
resized_image = cv2.resize(image, (target_width, target_height))
```

```python
# Visualize the original and resized images
plt.figure(figsize=(10, 5))

# Plot the original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Plot the resized image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.axis('off')

plt.show()
```

```python
# Define constant values for addition
constants = [-500, -100, 0,500,100]  # Add different constant values to the image

# Apply addition operation with different constant values and visualize each resulting image
plt.figure(figsize=(15, 5))  # Set the figure size for plotting

for i, const in enumerate(constants):
    # Perform addition operation
    result_image = cv2.add(resized_image, const)

    # Plot the image
    plt.subplot(1, len(constants), i + 1)  # Create subplots for each resulting image
    plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for proper visualization
    plt.title(f'Image with Constant {const}')  # Set title with the constant value
    plt.axis('off')  # Turn off axis for cleaner visualization
plt.suptitle('Addition Operation with Different Constants')  # Set the main title for all subplots
plt.tight_layout()
plt.show()  # Show the plot with all resulting images
```

```python
# Define constant values for subtraction
constants = [-500, -100, 0, 500, 100]  # Add different constant values to the image

# Apply subtraction operation with different constant values and visualize each resulting image
plt.figure(figsize=(15, 5))  # Set the figure size for plotting

for i, const in enumerate(constants):
    # Perform subtraction operation
    result_image = cv2.subtract(resized_image, const)  # Use subtract function 

    # Plot the image
    plt.subplot(1, len(constants), i + 1)  # Create subplots for each resulting image
    plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for proper visualization
    plt.title(f'Image with Constant {const}')  # Set title with the constant value
    plt.axis('off')  # Turn off axis for cleaner visualization
plt.suptitle('Subtraction Operation with Different Constants')  # Set the main title for all subplots
plt.tight_layout()
plt.show()  # Show the plot with all resulting images
```

```python
# Define scaling values for multiplication
scaling_values = [5, 2, 1, 0.5, 0.1]  # Multiply the image with different scaling values

# Apply multiplication operation with different scaling values and visualize each resulting image
plt.figure(figsize=(15, 5))  # Set the figure size for plotting

for i, scale in enumerate(scaling_values):
    # Perform multiplication operation using NumPy
    result_image = (resized_image * scale).astype(np.uint8)

    # Plot the image
    plt.subplot(1, len(scaling_values), i + 1)  # Create subplots for each resulting image
    plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for proper visualization
    plt.title(f'Image with Scaling Factor {scale}')  # Set title with the scaling value
    plt.axis('off')  # Turn off axis for cleaner visualization

plt.suptitle('Multiplication Operation with Different Scaling Factors')  # Set the main title for all subplots
plt.tight_layout()
plt.show()  # Show the plot with all resulting images
```

```python


# Invert the image
inverted_image = 255 - resized_image  # Subtract pixel values from 255 to invert the image

# Convert BGR to RGB for proper visualization
inverted_image_rgb = cv2.cvtColor(inverted_image, cv2.COLOR_BGR2RGB)

# Plot the original and inverted images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(inverted_image_rgb)
plt.title('Inverted Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
